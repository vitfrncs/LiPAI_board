1) Escolha um novo dataset e construa um programa que calcule as medidas estatísticas. Procure usar as funções apresentadas nessa atividade, assim como, outras que estão disponíveis nas bibliotecas do Python

```python
# abrindo data set
import pandas as pd
import numpy as np
import seaborn as sns
dataset = pd.read_csv("data/winequality-red.csv", sep=';')
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
# mostrando medidas estaticias vistas nas aulas
for coluna in dataset.columns:
    print(f"Coluna: {coluna} ")
    print(f"Media: {dataset[coluna].mean():.2f}")
    print(f"Mediana: {dataset[coluna].median():.2f}")
    print("Desvio padrão: %2.f " % dataset[coluna].std())
    print("Variancia: %2.f" % dataset[coluna].var())
    print("Minimo: %2.f" % dataset[coluna].min())
    print("Máximo: %2.f " % dataset[coluna].max())
    print("\n","-" * 40, "\n")
```

2) Use o dataset escolhido e mostre por meio das representações gráficas (histogramas) as principais características. Procure usar as funções apresentadas nesta atividade.

```python
print("\n=== 2) HISTOGRAMAS DAS PRINCIPAIS CARACTERÍSTICAS ===")
import matplotlib.pyplot as plt

# o data base tem 12 colunas, então criei uma figura com 3 linhas e 4 colunas 
fig, axes = plt.subplots(3, 4, figsize=(16, 12))

fig.suptitle('Histogramas das Características do Dataset de Vinhos', fontsize=16)

# copiando colunas para uma lista
columns = dataset.columns

# plotar histogramas
for i, col in enumerate(columns):
    row = i // 4
    col_idx = i % 4
    
    axes[row, col_idx].hist(dataset[col], bins=30, alpha=0.7, color='purple', edgecolor='black')
    axes[row, col_idx].set_title(f'{col}', fontsize=10)
    axes[row, col_idx].set_xlabel('Valores')
    axes[row, col_idx].set_ylabel('Frequência')
    axes[row, col_idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

5) Compare os resultados obtidos usando um gráfico box plot. Que atributos possuem distribuições mais simétricas? Há a presença de valores extremos, denominados de outliers, para os atributos?

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 4, figsize=(16, 12))
axes = axes.flat  # iterator
names = dataset.columns  # todas as colunas

for ax, col in zip(axes, names):
    ax.set_title(col)
    ax.boxplot(dataset[col])

plt.tight_layout()
plt.show()


# Como demostrado nos histogramas, há diversos valores outliers. As distribuições mais simetricas são: citric acid e alcohol
# Isso para meio errado não sei. Mas pelos histogramas até que faz sentido talvez?
```

6) No dataset escolhido selecione intervalos de amostragem (10%, 30% e 50% - Progressiva) e faça análises sobre os dados em relação as métricas estatísticas. O que você pode observar sobre os subconjuntos em relação as métricas estatísticas com a nova composição dos dados?

```python
amostragem = [0.1, 0.3, 0.5, 0.7, 1]

for p in amostragem:
    print(f'\nAnálise estatística de {p*100:.0f}% dos dados\n' + 50*'-')
    
    n_primeiras = int(len(dataset) * p) # pega o tamanho lentgh * p em amostragem
    amostra = dataset.iloc[:n_primeiras]

    print(amostra.describe())
    print("")

```

À medida que o tamanho da amostra cresce, as métricas estatísticas tendem a se estabilizar e a refletir melhor as do conjunto completo de dados. Em contrapartida, amostras pequenas são mais vulneráveis a valores extremos, como outliers, o que pode distorcer a percepção da verdadeira distribuição.
